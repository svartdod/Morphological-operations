```python
#Load Imagee
image=cv2.imread('tools_blackandwhite.png',0)
plt.imshow(image,cmap="gray")
#print(image.shape)
```

```python
#create structring element
size = 3
edge_padding=int((size-1)/2)
```

```python
#Morphology operation 
def MorphologyOperation(image,size,operation="Erode"):
  height, width = image.shape
  Karnel = np.ones((size,size),dtype=np.uint8)
  Result=np.zeros((height,width),dtype=np.uint8)
  for i in range (size,height-size):
    for j in range (size,width-size):
      if(operation is 'Erode'):
        Result[i,j]=np.min(np.multiply(image[i-size:i+size+1,j-size:j+size+1],Karnel))
      if (operation is 'Dilatation'):
        Result[i,j]=np.max(np.multiply(image[i-size:i+size+1,j-size:j+size+1],Karnel))

  return Result
```

```python
#Erode
Image_results =MorphologyOperation(image,edge_padding)
plt.imshow(Image_results,cmap="gray")
```

```python
#Dilitation
Image_results =MorphologyOperation(image,edge_padding,'Dilatation')
plt.imshow(Image_results,cmap="gray")
```

```python
#Opening
def Opening(image,size):
  Erode=MorphologyOperation(image,size)
  Dilate=MorphologyOperation(Erode,size,'Dilatation')
  return Dilate
```

```python
#Opening Results
Image_results=Opening(image,edge_padding)
plt.imshow(Image_results,cmap="gray")
```

```python
def Closing(image,size):
  Dilate=MorphologyOperation(image,size,'Dilatation')
  Erode=MorphologyOperation(Dilate,size)
  return Erode
```

```python
Image_results=Closing(image,edge_padding)
plt.imshow(Image_results,cmap="gray")
```
